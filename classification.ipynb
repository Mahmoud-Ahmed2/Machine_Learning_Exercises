```python
from urllib.request import urlretrieve
import pandas as pd
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
link = "https://assets.datacamp.com/production/repositories/5981/datasets/9e61071df2dd3811bacee34d670364c313674156/churn_modified.csv"
churn_df = pd.read_csv(urlretrieve(link, 'churn_modified.csv')[0])
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Import the module
from sklearn.model_selection import train_test_split

X = churn_df.drop("churn", axis=1).values
y = churn_df["churn"].values

# Split into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X , y, test_size=0.25, random_state= 21, stratify= y)
knn = KNeighborsClassifier(n_neighbors=6)

# Fit the classifier to the training data
knn.fit(X_train , y_train)
# Print the accuracy
print(knn.score(X_test , y_test))
```

```python
train_accuracy = {}
test_accuracy = {}
neighbors = np.arange(1,30)
for neighbor in neighbors:
    knn = KNeighborsClassifier(n_neighbors= neighbor)
    knn.fit(X_train, y_train)
    train_accuracy[neighbor] = knn.score(X_train, y_train)
    test_accuracy[neighbor] = knn.score(X_test, y_test)
```

```python
plt.figure (figsize=(8,6))
plt.plot(neighbors , train_accuracy.values())
plt.plot(neighbors , test_accuracy.values())
plt.legend()
plt.title('KNN : Varying Number of Neighbors')
plt.xlabel('Number of neighbors')
plt.ylabel('Accuracy')
plt.show()
```
